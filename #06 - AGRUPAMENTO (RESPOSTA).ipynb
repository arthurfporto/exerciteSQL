 ### Com base no esquema apresentado acima, escreva o código SQL para recuperar as seguintes informações no bando de dados.

<b> A) </b> A quantidade de professores por título

_Resposta:_

```python
%%sql 
SELECT professorTitulo, COUNT(*) AS quantidade
FROM professor
GROUP BY professorTitulo;
```

<b> B) </b> Nome dos estudantes que são conselheiros e a quantidade de estudantes que ele aconselha. 

_Resposta:_

```python
%%sql 
SELECT conselheiro.estudanteNome, COUNT(*)
FROM estudante AS aconselhado, estudante AS conselheiro, curso
WHERE 
    aconselhado.FK_ESTUDANTE_estudanteMatricula = conselheiro.estudanteMatricula AND
    conselheiro.FK_CURSO_cursoCodigo = curso.cursoCodigo
GROUP BY conselheiro.estudanteMatricula
ORDER BY conselheiro.estudanteNome
```

<b> C) </b> Nome do curso e a soma do orçamenato dos projetos que os seus estudantes participam, ordenados pelo valor da soma dos orçamentos de projeto. Motre apenas os cursos onde a soma é menor que R$ 50.000,00.

_Resposta:_

```python
%%sql

SELECT cursoNome, SUM(projetoOrcamento) soma
FROM curso, estudante, estudante_projeto, projeto
WHERE 
    cursoCodigo = FK_CURSO_cursoCodigo AND
    estudante.estudanteMatricula = estudante_projeto.FK_ESTUDANTE_estudanteMatricula AND
    FK_PROJETO_projetoNumero = projetoNumero
GROUP BY cursoCodigo
HAVING soma < 50000
ORDER BY soma
```

<b> D) </b> Quantos cursos existem em cada departamento, ordenados pelo nome do departamento.

_Resposta:_

```python
%%sql

SELECT departamentoNome, COUNT(*)
FROM 
    departamento JOIN  curso ON departamentoCodigo = FK_DEPARTAMENTO_departamentoCodigo
GROUP BY departamentoCodigo
ORDER BY departamentoNome
```

<b> E) </b> A média de carga horária dos professores por departamento. 

_Resposta:_

```python
%%sql

SELECT departamentoNome, AVG(cargaHoraria)
FROM professor_departamento JOIN departamento ON FK_DEPARTAMENTO_departamentoCodigo = departamentoCodigo
GROUP BY FK_DEPARTAMENTO_departamentoCodigo
```

<b> F) </b> A carga horária total dos professores que são coordenadores de departamento. Ordene os resultados pela soma da carga horária.

_Resposta:_

```python
%%sql

SELECT professorNome, SUM(cargaHoraria)
FROM 
    departamento d JOIN professor ON d.FK_PROFESSOR_professorMatricula = professorMatricula
    JOIN professor_departamento pd ON professorMatricula = pd.FK_PROFESSOR_professorMatricula
GROUP BY pd.FK_PROFESSOR_professorMatricula
```

<b> G) </b> O nome do coordenador, titulo do projeto e data inicial do projeto mais antigo deste professor. Mostre os resultados apenas para os professores que tenham primeiros projetos iniciados no ano de 2018. Ordene pela data inicial do projeto.

_Resposta:_

```python
%%sql

SELECT professorNome, projetoTitulo, MIN(projetoDataInicio) data
FROM professor JOIN projeto ON FK_PROFESSOR_professorMatricula = professorMatricula
GROUP BY professorMatricula
HAVING data >= '2018-01-01' AND data <= '2018-12-31'
ORDER BY data
```
