 ### Com base no esquema apresentado acima, escreva o código SQL para recuperar as seguintes informações no bando de dados.

##### Use apenas o Join explícito!

<b> A) </b> Título dos projetos e nome do coordenador.

_Resposta:_

```python
%%sql 
SELECT projetoTitulo, professorNome 
FROM 
    projeto JOIN professor ON professorMatricula = FK_PROFESSOR_professorMatricula 
ORDER BY projetoTitulo
```

<b> B) </b> Nome de todos os estudantes e nome do conselheiro (caso possuam). Apresente o resultado em ordem crescente pelo nome dos estudantes _Obs.: Utilize o 'ALIAS' para facilitar a identificação do conselheiro e aconselhado_.

_Resposta:_

```python
%%sql 
SELECT aconselhado.estudanteNome AS estudante, conselheiro.estudanteNome AS conselheiro
FROM 
    estudante AS aconselhado LEFT JOIN estudante AS conselheiro
    ON aconselhado.FK_ESTUDANTE_estudanteMatricula = conselheiro.estudanteMatricula
ORDER BY aconselhado.estudanteNome 
```

<b> C) </b> Nome dos professores, em ordem alfabética, que não são coordenadores de projeto.


_Resposta:_

```python
%%sql 
SELECT professorNome, projetoTitulo
FROM 
    professor LEFT JOIN projeto
    ON professorMatricula = FK_PROFESSOR_professorMatricula
WHERE
    FK_PROFESSOR_professorMatricula IS NULL
ORDER BY professorNome
```

<b> D) </b> Nome do curso e título dos projetos que os seus estudantes participam, ordenados pelo nome do curso. _Obs.: Apresente os resultados sem duplicatas_.


_Resposta:_

```python
%%sql

SELECT DISTINCT cursoNome, projetoTitulo
FROM 
    curso JOIN  estudante on cursoCodigo = FK_CURSO_cursoCodigo 
    JOIN estudante_projeto on estudante.estudanteMatricula = estudante_projeto.FK_ESTUDANTE_estudanteMatricula
    JOIN projeto ON FK_PROJETO_projetoNumero = projetoNumero
ORDER BY cursoNome
```

<b> E) </b> Nome dos alunos do mesmo curso do estudante 'Ilídio Canedo Tuna', em ordem alfabética. 

```python
%%sql

SELECT colega.estudanteNome
FROM 
    estudante AS colega JOIN estudante AS ilidio 
    ON ilidio.FK_CURSO_cursoCodigo = colega.FK_CURSO_cursoCodigo
WHERE 
    ilidio.estudanteNome = 'Ilídio Canedo Tuna'
ORDER BY colega.estudanteNome
```

<b> F) </b> Título dos projetos e nome coordenador ordenado pelo título do projeto. 

```python
%%sql

SELECT projetoTitulo, professorNome
FROM
    projeto JOIN professor ON  projeto.FK_PROFESSOR_professorMatricula = professor.professorMatricula
    JOIN professor_departamento AS pd ON professor.professorMatricula = pd.FK_PROFESSOR_professorMatricula
    JOIN departamento ON pd.FK_DEPARTAMENTO_departamentoCodigo = departamento.departamentoCodigo
WHERE
    departamentoNome = 'Computação'
```
